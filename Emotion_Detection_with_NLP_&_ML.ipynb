```python
import string

def remove_punctuation(text):
  for punctuation in string.punctuation:
    text = text.replace(punctuation, '')
  return text
```

```python
def remove_emoji(text):
  new = ""
  for i in text:
    if i.isascii():
      new+=i
  return new

df['text'] = df['text'].apply(remove_emoji)
df['text']
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')

stop_words = set(stopwords.words('english'))

```

```python
def remove_stopwords(text):
  word_tokens = word_tokenize(text)
  cleaned_text = []
  for i in word_tokens:
    if i not in stop_words:
      cleaned_text.append(i)
  return ' '.join(cleaned_text)

df['text'] = df['text'].apply(remove_stopwords)
```

```python
x = df['text']
y = df['emotion']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer

tf_idf_vectorizer = TfidfVectorizer()
bow_vectorizer = CountVectorizer()

X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)

X_train_tf_idf = tf_idf_vectorizer.fit_transform(X_train)
X_test_tf_idf = tf_idf_vectorizer.transform(X_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

nb_bow_model = MultinomialNB()
nb_bow_model.fit(X_train_bow, y_train)
```

```python
pred_nb_bow = nb_bow_model.predict(X_test_bow)
print(accuracy_score(y_test, pred_nb_bow))
```

```python
nb_tf_idf_model = MultinomialNB()
nb_tf_idf_model.fit(X_train_tf_idf, y_train)
```

```python
pred_nb_tf = nb_tf_idf_model.predict(X_test_tf_idf)
print(accuracy_score(y_test, pred_nb_tf))
```
